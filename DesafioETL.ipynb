## **E**xtract

Extraia a lista de IDs de usuário a partir do arquivo CSV. Para cada ID, faça uma requisição GET para obter os dados do usuário correspondente.

```python
import pandas as pd

# Ler o dataset
dataset = pd.read_csv("/content/drive/MyDrive/bootcamp_santander/Iris-2-1.csv").dropna()
# Inserir em um dataframe
df = pd.DataFrame(dataset)
```

## **T**ransform

Utilize a API do OpenAI GPT-4 para gerar uma mensagem de marketing personalizada para cada usuário.

```python
# Importar o método PCA da biblioteca scikit-learn
from sklearn.decomposition import PCA

# Separar os atributos previsores
X = df.iloc[:,0:-1]

# Separar o atributo alvo
Y = df.iloc[:,-1]

# Realizar o PCA entre os atributos previsores
pca = PCA(n_components=3)
components = pca.fit_transform(X)
total_var = pca.explained_variance_ratio_.sum() * 100
```

## **L**oad

Atualize a lista de "news" de cada usuário na API com a nova mensagem gerada.

```python
# Importar a biblioteca numpy
import numpy as np

# Converter a variável Y para uma matriz NumPy e ajustar as dimensões
Y = np.array(Y).reshape(-1, 1)

# Concatenar a matriz components com o vetor Y
arrayPCA = np.concatenate((components, Y), axis=1)

# Inserir a mtriz arrayPCA em um DataFrame
dfPCA = pd.DataFrame(arrayPCA)

# Carregar os dados transformados em uma planilha tradicional do Excel (.xlsx)
dfPCA.to_excel('/content/drive/MyDrive/bootcamp_santander/IrisPCA.xlsx', index=False)
```
